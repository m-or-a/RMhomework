```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
x = np.linspace(-4, 4, 200)
```

Нормальное распределение

```python
plt.plot(x, stats.norm.pdf(x, 0, 1))
plt.title('Плотность вероятности')
```

```python
plt.plot(x, stats.norm.cdf(x, 0, 1))
plt.title('Распределение')
```

Экспоненциальное распределение

```python
plt.plot(x, stats.expon.pdf(x, 0, 1))
plt.title('Плотность вероятности')
```

```python
plt.plot(x, stats.expon.cdf(x, 0, 1))
plt.title('Распределение')
```

Равномерное распределение

```python
plt.plot(x, stats.uniform.pdf(x, 0, 1))
plt.title('Плотность вероятности')
```

```python
plt.plot(x, stats.uniform.cdf(x, 0, 1))
plt.title('Распределение')
```

Биномиальное распределение

```python
n, p = 12, 0.4
x = np.arange(0, n+1)

plt.scatter(x, stats.binom.pmf(x, n, p))
plt.title('Плотность вероятности')
```

```python
plt.plot(x, stats.binom.cdf(x, n, p))
plt.title('Распределение')
```

Распределение Пуассона

```python
mu = 3

plt.scatter(x, stats.poisson.pmf(x, mu))
plt.title('Плотность вероятности')
```

```python
plt.plot(x, stats.poisson.cdf(x, mu))
plt.title('Распределение')
```
